```python
# data visualization

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['average_rating'],bins=20)
plt.title('Distribution of average ratings')
plt.show()
```

```python
top_authors = df['authors'].value_counts().head(10)
sns.barplot(x=top_authors.values, y=top_authors.index)
plt.title('Top 10 authors by number of books')
plt.show()
```

```python
pub_trends = df['published_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=pub_trends.index, y=pub_trends.values)
plt.title('Number of books published per year')
plt.xlabel('Year')
plt.ylabel('Number of books')
plt.xticks(rotation=45)
plt.show()
```

```python
def simplify_category(category):
    if pd.isnull(category):
        return 'Unknown'
    if 'Fiction' in category:
        return 'Fiction'
    elif 'Non-Fiction' in category:
        return 'Non-Fiction'
    elif 'Science' in category or 'Technology' in category:
        return 'Science & Technology'
    elif 'History' in category:
        return 'History'
    else:
        return 'Other'
```

```python
df['simplified_category'] = df['categories'].apply(simplify_category)
```

```python
cat_counts = df['simplified_category'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=cat_counts.index, y=cat_counts.values)
plt.title('Distribution of books according to categories')
plt.xlabel('Category')
plt.ylabel('Number of books')
plt.xticks(rotation=45)
plt.show()
```

```python
df['num_pages'].fillna(0, inplace=True) 
df['average_rating'].fillna(0, inplace=True)
```

```python
df['rating_per_page'] = df.apply(lambda row: row['average_rating']/row['num_pages'] if row['num_pages']>0 else 0,axis=1)
```
